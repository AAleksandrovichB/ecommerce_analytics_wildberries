### Произовдим проверку текущих остатков на складе на премдмет пропроционального соотвествия ABC-анализу

```python
import pandas as pd
import numpy as np
# Загружаем таблицы о продажа и unit-экономике
sales = pd.read_csv(r'C:\Users\22\Desktop\Костя\Данные_чистые\sales.csv', sep=';', decimal=',')
unit_economics = pd.read_csv(r'C:\Users\22\Desktop\Костя\Данные_чистые\unit_economics.csv', sep=';')
```

```python
# 1. Берём необходимые столбцы из таблицы unit_economics
ue = unit_economics[['sku', 'abc', 'cost_price']]
# 2. Объединяем их с таблицей sales
sales_merged = sales.merge(ue, on='sku', how='inner')
# Проводим типы
sales_merged['stock'] = pd.to_numeric(sales_merged['stock'], errors='coerce')
sales_merged['cost_price'] = pd.to_numeric(sales_merged['cost_price'], errors='coerce')
# 3. Считаем суммарный остаток по артикулу
df_stock = sales_merged.groupby(['sku', 'abc', 'cost_price'])['stock'].sum().reset_index()
# 4. Считаем "замороженный капитал" 
df_stock['frozen_capital'] = df_stock['stock'] * df_stock['cost_price']
# 5. Суммируем по группам ABC и считаем доли
result = df_stock.groupby('abc')['frozen_capital'].sum().reset_index()
result['frozen_pct'] = result['frozen_capital'] / result['frozen_capital'].sum() * 100
result
```

```python
result.to_csv(r'C:\Users\22\Desktop\Костя\Данные_чистые\frozen_capital.csv', index=False, sep=';', decimal=',')
```
